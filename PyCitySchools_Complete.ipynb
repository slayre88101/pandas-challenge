---Based upon the above data, one would conclude that Charter schools have a higher degree of learning and set students up for scholastic success. However, the data is skewed as Charter schools tend to have a much smaller student body than those of District schools. Larger schools, typical of district schools, have a much lower overall passing rate than smaller schools. Interestingly, the schools with more money per student have a lower overall passing rate. This comes as a surprise as one would expect wealthier schools to have a higher degree of learning as they can afford more teaching resources. 

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete["School ID"].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["Student ID"].nunique()
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary_df = pd.DataFrame([[school_count,student_count,total_budget,average_math_score,average_reading_score,passing_math_percentage, passing_reading_percentage, overall_passing_rate]], columns=['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score','Passing Math Percentage','Passing Reading Percentage', 'Overall Passing Rate'])

# Formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{0:.2f}".format)
district_summary_df["Passing Math Percentage"] = district_summary_df["Passing Math Percentage"].map("{:.2f}".format)
district_summary_df["Passing Reading Percentage"] = district_summary_df["Passing Reading Percentage"].map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["Overall Passing Rate"] = district_summary_df["Overall Passing Rate"].map("{:.2f}".format)
# Display the DataFrame
district_summary_df
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data['type'].value_counts()
school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete["school_name"].value_counts()
per_school_counts

```

```python
# Calculate the total school budget and per capita spending per school
budget = school_data_complete['budget']
budget = school_data_complete.groupby(['school_name']).mean(numeric_only=True)
budget_per_capita = budget['budget'] / per_school_counts
total_budget=budget['budget']
total_budget

```

```python
# Calculate the average test scores per school
math_per_school = school_data_complete['math_score']
math_per_school = school_data_complete.groupby(['school_name']).mean(numeric_only=True)
math_per_school= math_per_school['math_score']

reading_per_school = school_data_complete['reading_score']
reading_per_school = school_data_complete.groupby(['school_name']).mean(numeric_only=True)
reading_per_school= reading_per_school['reading_score']
```

```python
# Calculate the number of students per school with math scores of 70 or higher
school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
school_passing_math=school_passing_math.groupby(["school_name"]).size()
school_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
school_passing_reading= school_passing_reading.groupby(["school_name"]).size()
school_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
#Use the provided code to calculate the passing rates
per_school_passing_math = school_passing_math / per_school_counts * 100
per_school_passing_reading = school_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
overall_passing_rate

```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({'Number of Students': per_school_counts,
                                   'Total Budget Per School':total_budget,
                                   "Budget per Student":budget_per_capita,
                                   "Average Math Score":math_per_school,
                                   "Average Reading Score":reading_per_school,
                                   'Students Passing Math':school_passing_math,
                                   'Students Passing Reading':school_passing_reading,
                                   'Math Passing Rates':per_school_passing_math,
                                   'Reading Passing Rates':per_school_passing_reading,
                                   'Students Passing Both':school_students_passing_math_and_reading,
                                   'Overall Passing Rate':overall_passing_rate
                                   }) 

# Formatting
per_school_summary["Budget per Student"] = per_school_summary["Budget per Student"].map("${:,.2f}".format)
per_school_summary["Total Budget Per School"] = per_school_summary["Total Budget Per School"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# Establish the bins 
bins = [0, 585, 630, 645, 680]
group_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.

school_spending_df["Spending Ranges (Per Student)"] = pd.cut(budget_per_capita,bins,labels=group_labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Math Passing Rates"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Reading Passing Rates"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Passing Rate"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame ({"Average Math Score":spending_math_scores,
                                      "Average Reading Score":spending_reading_scores,
                                      "% Passing Math":spending_passing_math,
                                      "% Passing Reading":spending_passing_reading,
                                      '% Overall Passing':overall_passing_spending
})

# Display results
spending_summary
```
